   üëâ Une compagnie d'assurance automobile re√ßoit en moyenne <span style="background-color: #D3D3E9; color: #000000;"><b>1000 demandes de remboursement</b></span> chaque mois pour les sinistres automobiles. La probabilit√© qu'une demande de remboursement soit accept√©e en raison d'un accident est de <span style="background-color: #D3D3E9; color: #000000;"><b>0.25</b></span>. En cas de sinistre accept√©, <span style="background-color: #D3D3E9; color: #000000;"><b>la compagnie rembourse en moyenne 3000 ‚Ç¨</b></span>. On suppose que chaque demande de remboursement est <span style="background-color: #D3D3E9; color: #000000;"><b>ind√©pendante des autres</b></span> et suit une distribution binaire (binomiale) avec une probabilit√© d'acceptation de 0.25.
</div>

## <div style=" background-color: #4A4A90; color:white; padding: 10px; line-height: 1.5;"> Importer les biblioth√®ques üìö
</div>

```python
# Importer la fonction "comb" depuis le module math.

from math import comb
```

<div style="font-family: Segoe UI; line-height: 2; color: #000000; text-align: justify">
    
   <span style="background-color: #D3D3E9; color: #000000;"><b>"Comb"</b></span> est une fonction pour calculer <span style="background-color: #D3D3E9; color: #000000;"><b>le coefficient binomial.</b></span>
</div>

$$ C(n, k) = \frac{n!}{k!(n-k)!} $$


```python
# Le nombre de combinaisons avec 5 tirages et avoir exactement 2 √©l√©ments.

comb(5, 2)
```

<div style="font-family: Segoe UI; line-height: 2; color: #000000; text-align: justify">
    
   Il y a 10 combinaisons possibles.
</div>

```python
# La fonction de la loi binomiale :

def binomial_formula(n, k, p):
    return comb(n, k) * (p**k) * ((1-p)**(n-k)) 
```

$$ C(n, k) \cdot p^k \cdot (1-p)^{n-k} $$

    </div>

```python
# Cr√©er une liste qui va contenir toutes les probabilit√©s sup√©rieures ou √©gales √† 250
plus_grand_que_250 = []

# Avec une boucle for, calculer chaque probabilit√© sup√©rieur √† 250
for i in range(250, 1_001):
    # Enregistrer chaque probabilit√© dans la liste
    plus_grand_que_250.append(binomial_formula(1_000, i, 0.25))

# Faire la somme de la liste

round(sum(plus_grand_que_250), 4) *100
```

## <div style=" background-color:#4A4A90; color:white; padding: 12px; line-height: 1.5;">Approche avec la simulation de Monte Carlo</div>

<div style="font-family: Segoe UI; line-height: 2; color: #000000; text-align: justify">
La simulation de Monte Carlo permet <span style="background-color: #D3D3E9; color: #000000;"><b>d‚Äôestimer de mani√®re empirique un √©v√©nement.</b></span>

<u>Comment allons-nous proc√©der ?</u>

Nous allons <span style="background-color: #D3D3E9; color: #000000;"><b>g√©n√©rer de mani√®re al√©atoire un nombre de r√©clamations accept√©es</b></span> en utilisant la distribution binomiale avec une probabilit√© de 0.25.


<b><u>Par exemple :</u></b> Nous g√©n√©rons 233 r√©clamations qui sont accept√©es.
    
Nous calculons le co√ªt total des r√©clamations accept√©es <span style="background-color: #D3D3E9; color: #000000;"><b>en multipliant</b></span> le nombre de r√©clamations accept√©es par le montant moyen de remboursement, <span style="background-color: #D3D3E9; color: #000000;"><b>soit : 233 * 3 000 = 699 000.</b></span>

Avec 233 r√©clamations accept√©es, l‚Äôassurance paie 699 000‚Ç¨.
Nous avons r√©alis√© 1 it√©ration, <span style="background-color: #D3D3E9; color: #000000;"><b>nous allons produire cette simulation 10 000 fois</b></span>, et <span style="background-color: #D3D3E9; color: #000000;"><b>regarder combien de fois nous d√©passons 750 000‚Ç¨.</b></span>
Si sur 10 000 it√©rations, <span style="background-color: #D3D3E9; color: #000000;"><b>5 000 sont sup√©rieurs √† 750 000‚Ç¨</b></span>, alors la probabilit√© de d√©passer 750 000‚Ç¨ est de <span style="background-color: #D3D3E9; color: #000000;"><b>50%.</b></span>
    
<u>Comment allons-nous g√©n√©rer de mani√®re al√©atoire un nombre de r√©clamations accept√©es ?</u>
    
Allons-nous choisir un nombre al√©atoire entre 0 et 10 000 ?
Presque, le probl√®me de choisir un nombre entre 0 et 10 000 est que la distribution est uniforme. Nous, <span style="background-color: #D3D3E9; color: #000000;"><b>nous avons une distribution binomiale</b></span>. Il y a une plus grande probabilit√© que l‚Äôassurance <span style="background-color: #D3D3E9; color: #000000;"><b>accorde le remboursement de 200 r√©clamations que de 0.</b></span> Il faut que le hasard suive la distribution binomiale.
</div>

```python
# Cr√©er la distribution binomiale
vraie_distribution = []
for i in range(1001):
    vraie_distribution.append(binomial_formula(1000, i, 0.25))
```

    <br>
    Donc le nombre que nous devons g√©n√©rer al√©atoirement doit <span style="background-color: #D3D3E9; color: #000000;"><b> suivre cette distribution.</b></span>
</div>



```python
# Renvoie un nombre al√©atoire en suivant cette distribution binomiale avec 1 000 tirages
# et 25% de chances que l'√©v√©nement se produise
np.random.seed(42)

nbr_reclamation_acceptee = np.random.binomial(1_000, 0.25)
cout = nbr_reclamation_acceptee * 3_000
cout
```

```python
# Cr√©er une variable qui va mesurer le nombre de r√©clamations sup√©rieures √† 750 000e

plus_grand_que_750_000 = 0

if cout >= 750_000:
    plus_grand_que_750_000 +=1

plus_grand_que_750_000
```

```python
np.random.seed(42)

# Initialise la variable par 0
plus_grand_que_750_000 = 0

# Cr√©er une boucle for allant jusqu'√† 10 000
for i in range(10_000):
    # G√©n√®re un nombre al√©atoire de la distribution binomiale avec 1 000 tirages et une probabilit√© de 25%
    nbr_reclamation_acceptee = np.random.binomial(1_000, 0.25)
    # Mesure le co√ªt en fonction du nombre de r√©clamations accept√©es
    cout = nbr_reclamation_acceptee * 3_000
    # Si le co√ªt est sup√©rieur √† 750 000, nous ajoutons un "1"
    if cout >= 750_000:
        plus_grand_que_750_000 +=1
        
print(f"La probabilit√© est : {round((plus_grand_que_750_000/10_000)*100, 4)}%")
```

<div style="font-family: Segoe UI; line-height: 2; color: #000000; text-align: justify">
   La probabilit√© que les co√ªts d√©passent 750 000‚Ç¨ est de <span style="background-color: #D3D3E9; color: #000000;"><b>50,96%.</b></span> C‚Äôest-√†-dire que sur 10 000 simulations, <span style="background-color: #D3D3E9; color: #000000;"><b>5 096 fois</b></span> les co√ªts d√©passent 750 000‚Ç¨.
    <br>
    Cette probabilit√© est pratiquement <span style="background-color: #D3D3E9; color: #000000;"><b>la m√™me </b></span> qu‚Äôavec l‚Äôapproche classique √† quelques centi√®mes pr√®s.
